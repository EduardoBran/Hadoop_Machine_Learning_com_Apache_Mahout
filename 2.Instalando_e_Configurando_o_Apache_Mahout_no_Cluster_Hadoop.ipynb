# Instalando e Configurando o Apache Mahout no Cluster Hadoop

<br>

---

<br><br>

## Download (versão 13)

> https://archive.apache.org/dist/mahout/0.13.0/

<br>

### Descompactar

Descompactar com:

```bash
tar -xvf apache-mahout-disibution-0.13.0.tar.gz
```

<br>

### Mover Diretório

Mover diretório:

```bash
sudo mv apache-mahout-distribution-0.13.0 /opt/mahout
```

<br>

### Configurar Variável de Ambiente

Abrir arquivo .bashrch:

```bash
gedit .bashrc
```

Em .bashrc adicionar as seguintes linhas:

```bash
# Mahout
export MAHOUT_HOME=/opt/mahout
export PATH=$PATH:$MAHOUT_HOME/bin
```

Aplicar o source:

```bash
source .bashrc
```

<br>

### Testando

Digitar no terminal:

```bash
mahout
```
